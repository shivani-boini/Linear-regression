```python
data.drop(['Medical_History_32'],axis=1)
```

```python
data.dropna(subset=['Response'],inplace=True)
data.isnull().sum()
```

```python
# Identify the input columns (a list of column names)
input_cols = data.columns[0:127]
input_cols
```

```python
# Identify the name of the target column (a single string, not a list)
target_col = data.columns[127]
target_col
```

```python
inputs_df = data[input_cols].copy()
```

```python
targets = data[target_col]
targets.isnull().sum()
targets
```

```python
numeric_cols = inputs_df.select_dtypes(include=['int64','float64']).columns.tolist()
```

```python
categorical_cols = inputs_df.select_dtypes(include=["object"]).columns.tolist()
categorical_cols
```

```python
from sklearn.impute import SimpleImputer
```

```python
# 1. Create the imputer
imputer = SimpleImputer(strategy='mean')
```

```python
# 2. Fit the imputer to the numeric colums
imputer.fit(inputs_df[numeric_cols])
```

```python
# 3. Transform and replace the numeric columns
inputs_df[numeric_cols] = imputer.transform(inputs_df[numeric_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create the encoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python
# 2. Fit the encoder to the categorical colums
encoder.fit(inputs_df[categorical_cols])
```

```python
# Generate column names for each category
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
len(encoded_cols)
```

```python
# 4. Transform and add new one-hot category columns
inputs_df[encoded_cols] = encoder.transform(inputs_df[categorical_cols])
```

```python
inputs_df
```

```python
input_col=inputs_df.drop(['Product_Info_2','Id'],axis=1)

```

```python
input_col
```

```python
targets
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_inputs, val_inputs, train_targets, val_targets = train_test_split(input_col,
                                                                        targets,
                                                                        test_size=0.25,
                                                                        random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
# Fit the model using inputs and targets
x=model.fit(train_inputs,train_targets)
x
```

```python
from sklearn.metrics import mean_squared_error
```

```python
train_preds = model.predict(train_inputs)
```

```python
train_mse = mean_squared_error(train_preds, train_targets)
train_rmse = np.sqrt(train_mse)
```

```python
print('The RMSE loss for the training set is $ {}.'.format(train_rmse))
```

```python
val_preds = model.predict(val_inputs)
```

```python
val_mse = mean_squared_error(val_preds, val_targets)
val_rmse = np.sqrt(val_mse)
```

```python
print('The RMSE loss for the validation set is $ {}.'.format(val_rmse))
```

```python

 #Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,#here we willselect no of decicion trees to be built.
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Random search of parameters, using 3 fold cross validation,
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(train_inputs,train_targets)
```

```python
rf2_train_preds = rf_random.predict( train_inputs)
```

```python
rf2_train_mse = mean_squared_error(rf2_train_preds, train_targets)
rf2_train_rmse = np.sqrt(rf2_train_mse)
rf2_train_rmse
```
